<i>Matplotlib is your best friend</i>
<i>numpy is just for building data right now</i>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('ggplot')
```

```python
#define some basic data
x = np.arange(0,100)
y = [np.random.randint(x-x*0.5,x+2) for x in x]
y = np.array(y, dtype=int)
```

```python
#example of how to plot data in matplotlib
%matplotlib inline
plt.plot(x,y,linewidth=1)
```

```python
#or you could make different looking plot with styles. 
plt.style.use('dark_background')
plt.plot(x,y)
plt.show()
plt.style.use('fivethirtyeight')
plt.plot(x,y)
plt.show()
plt.style.use('seaborn-darkgrid')
plt.plot(x,y, linewidth=2)
```
